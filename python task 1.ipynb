```python
import requests

url = "https://catfact.ninja/facts?limit=5"
response = requests.get(url)
data = response.json()

# Extract facts
facts = [item['fact'] for item in data['data']]
print(facts)
```

```python
# Count words in each fact
word_counts = [len(fact.split()) for fact in facts]
print(word_counts)
```

```python
import matplotlib.pyplot as plt

# Plot bar chart
plt.bar(range(len(facts)), word_counts, color='skyblue')
plt.title("Word Count in Each Cat Fact")
plt.xlabel("Fact Number")
plt.ylabel("Word Count")
plt.xticks(range(len(facts)), [f"Fact {i+1}" for i in range(len(facts))])
plt.tight_layout()
plt.show()
```
